# Competition

```python
import pandas as pd
df = pd.read_excel('comp_sales.xlsx')
df.head(3)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

variables = ['Price', 'Price_of_AngelSoft']
for var in variables:
    plt.figure() 
    sns.regplot(x=var, y='Sales', data=df).set(title=f'Regression plot of {var} and Sales');
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = df[['Price','Price_of_AngelSoft']]
y = df['Sales']

regressor = LinearRegression()
regressor.fit(x, y)
intercept_ = regressor.intercept_
slope_ = regressor.coef_

print(f"intercept: {intercept_}")
print(f"slope: {slope_}")
```

# Cross Price Elasticity

**Using package retail_stats**

```python
from math import isclose
from retail_stats import elasticity

original_quantity = 200
new_quantity = 400

original_price = 1000
new_price = 1050
# (200 / 300) / (50 / 1025)
expected_ced = 13.66666666666666
ced = elasticity.calculate_cross_elasticity(original_quantity, 
                                            new_quantity, 
                                            original_price, 
                                            new_price)

assert isclose(expected_ced, ced)
```
