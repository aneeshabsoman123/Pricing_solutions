```python
import pandas as pd
```

```python
df = pd.read_excel('choice_data_new.xlsx')
df.head(10)
```

```python
df.describe()
```

```python

import statsmodels.api as sm
from statsmodels.genmod.families import Binomial
from statsmodels.genmod.families.links import logit

df = pd.get_dummies(df, columns=['phone_version','phone_color','camera_type'])

X = df.drop({'choice_flag'},axis=1)
y = df['choice_flag']

model = sm.GLM(y, X, family=Binomial(link=logit()))
```

```python
results = model.fit()
```

```python
print(results.summary())
```

The coefficients of each attribute is understood from the table

```python
coefficients = results.params
print(coefficients)
```
